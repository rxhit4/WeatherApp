```python
import requests as rq
import json
import os
from gtts import gTTS
from IPython.display import Audio, Video
```

```python
city = input('Enter city name \n')
a = rq.get (f'https://api.weatherapi.com/v1/current.json?key=49696d462dcd466ebe875528241511&q={city}')
#print(a.text)
weather = json.loads(a.text)
x = weather['current']['temp_c']
tts = gTTS (f'The current weather in {city} is {x} degree celcius')
tts.save('weather.mp3')
soundfile = 'weather.mp3'
Audio(soundfile, autoplay = True)
```
